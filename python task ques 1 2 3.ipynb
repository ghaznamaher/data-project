1. Largest Difference in List
Problem: Given a list of integers, return the largest difference between any two elements, where the larger element appears after the smaller one. If no such pair exists, return 0.
Example: Input: [1, 2, 90, 10, 110] Output: 109 Explanation: The largest difference is between 1 and 110


```python
def largest_diff(nums):
    if not nums or len(nums) < 2:
        return 0

    min = nums[0]
    max_diff = 0

    for num in nums[1:]:
        if num > min:
            max_diff = max(max_diff, num - min)
        else:
            min= min(min, num)

    return max_diff
```

```python
largest_diff([1, 2, 90, 10, 110])
```

Problem: Given a string, return the first non-repeating character. If there are no non-repeating characters, return None.
Example: Input: "aabcc" Output: 'b' Explanation: The first non-repeating character is 'b'.

```python
def non_repeating_char(x):
    char_count = {}


    for char in x:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    for char in x:
        if char_count[char] == 1:
            return char

    return None
```

```python
non_repeating_char("wwrrt")
```

Given two strings,write a function to check if one is a permutation of the other. Return True if they are permutations,False otherwise.


```python
def permutations(a1, a2):
    if len(a1) != len(a2):
        return False
    return sorted(a1) == sorted(a2)
```

```python
permutations("sara","aras")
```
